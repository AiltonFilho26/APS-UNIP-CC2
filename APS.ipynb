```python
import time
def crip(frase1):

    frase1 = (str(frase1).replace('a','@').replace('b','!').replace('c','1')
    .replace('ç','£').replace('d','#').replace('e','3').replace('f','$')
    .replace('g','¨').replace('h','%').replace('i','6').replace('j','.')
    .replace('k','[').replace('l','/').replace('m','?').replace('n',';')
    .replace('o','^').replace('p','&').replace('q','*').replace('r','{')
    .replace('s','|').replace('t','º').replace('u','>').replace('v','<')
    .replace('w','§').replace('x','=').replace('y','+').replace('z','_')
    .replace('ã','8').replace('á','I').replace('â','P').replace('à','S')
    .replace('ê','¢').replace('é','0').replace('è','¬').replace('ì','5')
    .replace('í','7').replace('î','¹').replace('ô','4').replace('ó','²')
    .replace('ò','³').replace('õ','-').replace('ù','ª').replace('û','°')
    .replace('ú','9').replace(' ','"'))
    print('Sua frase Criptografada: ', frase1)

    print('='*40)

    print('[1] Descriptografar esta frase')
    print('[2] Criptografar outra frase')
    print('[3] Descriptografar outra frase')
    print('[4] Finalizar programa')

    print('='*40)

    escolha = int(input("Sua escolha aqui: "))
    if (escolha == 1):
        print("Descriptografando...")
        time.sleep(5)
        des(frase1)

    elif(escolha == 2):
        frase1 = str(input('Qual frase deseja Criptografar? ')).lower()
        print("Criptografando...")
        time.sleep(5)
        crip(frase1)

    elif(escolha == 3):
        frase2 = str(input('Qual frase deseja Descriptografar? '))
        des(frase2)
    
    elif(escolha == 4):
        print("Fim do programa!")

def des(frase2):

    frase2 = (str(frase2).replace('@','a').replace('!','b').replace('1','c')
    .replace('£','ç').replace('#','d').replace('3','e').replace('$','f')
    .replace('¨','g').replace('%','h').replace('6','i').replace('.','j')
    .replace('[','k').replace('/','l').replace('?','m').replace(';','n')
    .replace('^','o').replace('&','p').replace('*','q').replace('{','r')
    .replace('|','s').replace('º','t').replace('>','u').replace('<','v')
    .replace('§','w').replace('=','x').replace('+','y').replace('_','z')
    .replace('8','ã').replace('I','á').replace('P','â').replace('S','à')
    .replace('¢','ê').replace('0','é').replace('¬','è').replace('5','ì')
    .replace('7','í').replace('¹','î').replace('4','ô').replace('²','ó')
    .replace('³','ò').replace('-','õ').replace('ª','ù').replace('°','û')
    .replace('9','ú').replace('"',' '))
    print('Sua frase Descriptografada: ', frase2.lower().capitalize())

    print('='*40)

    print('[1] Criptografar esta frase')
    print('[2] Descriptografar outra frase')
    print('[3] Criptografar outra frase')
    print('[4] Finalizar programa')

    print('='*40)

    escolha = int(input("Sua escolha aqui: "))
    if (escolha == 1):
        print("Criptografando...")
        time.sleep(5)
        crip(frase2)

    elif(escolha == 2):
        frase2 = str(input('Qual frase deseja descriptografar? '))
        print("Descriptografando...")
        time.sleep(5)
        des(frase2)
    
    elif(escolha == 3):
        frase1 = str(input('Qual frase deseja Criptografar? ')).lower()
        print("Criptografando...")
        time.sleep(5)
        crip(frase1)
    
    elif(escolha == 4):
        print("Fim do programa!")
        

print("="*60)
print("  Bem-Vindo ao programa de criptografia e descriptografia!  ")
print("="*60)

print('O que deseja? ')

print('  [1] Criptografar')
print('  [2] Descriptografar')

escolha = int(input("Sua escolha aqui: "))

if (escolha == 1):
    frase1 = str(input('Qual frase deseja criptografar? ')).lower()
    print("Criptografando...")
    time.sleep(5)
    crip(frase1)

elif(escolha == 2):
    frase2 = str(input('Qual frase deseja descriptografar? '))
    print("Descriptografando...")
    time.sleep(5)
    des(frase2)
```
